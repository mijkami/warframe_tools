# Imports

```python
import requests
from IPython.display import JSON
import json
import pandas as pd

from bs4 import BeautifulSoup
import re
```

# Request API

```python
page_name = 'Damage'
```

```python
url = f"https://warframe.fandom.com/api.php?action=parse&page={page_name}&prop=text&format=json"

response = requests.get(url)
```

```python
# Vérifier si la requête a réussi
if response.status_code == 200:
    # Charger les données JSON
    data = response.json()
    text_data = data.get('parse', {}).get('text', {}).get('*', '')

    # Créer un DataFrame
    # Note: Vous devrez adapter cette partie en fonction de la structure des données
    df = pd.DataFrame({'text_data': [text_data]})

    # Afficher le DataFrame
    print(df)
else:
    print("La requête a échoué avec le code d'état :", response.status_code)
```

# Check Data + cleaning + export html

```python
data = response.json()
```

```python
JSON(data)
```

```python
data['parse']['text']
```

```python
text_content = data['parse']['text']
text_string = json.dumps(text_content)

# Suppression reliquats début et fin de texte JSON
cleaned_text = text_string.replace('{"*": "', '').replace('"}', '').strip()

# Supprimer les antislashs indésirables
cleaned_text = re.sub(r'\\"', '"', cleaned_text)

export_filename = f'wiki_{page_name}.html'

soup = BeautifulSoup(cleaned_text, 'html.parser')
pretty_html = soup.prettify()

with open(export_filename, 'w', encoding='utf-8') as file:
    file.write(pretty_html)

print(f"Le contenu a été exporté dans {export_filename}")
```

# Récupération paragraphe descriptif

```python
import requests
from bs4 import BeautifulSoup

# URL de la page web
url = "https://wiki.warframe.com/w/Damage"

# Faire une requête GET pour obtenir le contenu de la page
response = requests.get(url)

if response.status_code == 200:
    # Stocker le contenu HTML dans une variable
    html_content = response.content

    # Analyser le contenu HTML avec BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Trouver la div avec l'ID "mw-content-text"
    mw_content_div = soup.find('div', id='mw-content-text')

    if mw_content_div:
        # Trouver tous les h2
        h2_tags = mw_content_div.find_all('h2')

        if len(h2_tags) >= 2:
            # Trouver les paragraphes entre le premier et le deuxième h2
            first_h2 = h2_tags[0]
            second_h2 = h2_tags[1]

            # Parcourir tous les éléments entre les deux h2
            paragraphs = []
            current_element = first_h2.find_next()
            while current_element and current_element != second_h2:
                if current_element.name == 'p':
                    paragraphs.append(current_element.get_text(strip=True))
                current_element = current_element.find_next()

            # Joindre les paragraphes en une seule chaîne de texte
            text_content = '\n'.join(paragraphs)

            # Afficher le contenu extrait
            print("Contenu extrait :")
            print(text_content)

            # Nom du fichier d'exportation
            export_filename = 'extracted_content.txt'

            # Écrire le contenu texte dans un fichier
            with open(export_filename, 'w', encoding='utf-8') as file:
                file.write(text_content)

            print("\n")
            print(f"Le contenu texte a été exporté dans {export_filename}")
        else:
            print("Moins de deux balises h2 trouvées.")
    else:
        print("Aucune div avec l'ID 'mw-content-text' trouvée.")
else:
    print("La requête a échoué avec le code d'état :", response.status_code)
```

# Consolidation code

```python
import requests
import os
from bs4 import BeautifulSoup

def extract_and_save_content(keywords, print_content=True):
    base_url = "https://wiki.warframe.com/w/"
    export_dir = "exports"

    # Créer le dossier d'exportation s'il n'existe pas
    if not os.path.exists(export_dir):
        os.makedirs(export_dir)

    for keyword in keywords:
        # Ligne de séparation avant le traitement de chaque page
        print(f"\n{'-----' * 10}")
        print(f"Traitement de la page : {keyword}")
        print(f"{'-----' * 10}\n")

        # Construire l'URL complète
        url = base_url + keyword

        # Faire une requête GET pour obtenir le contenu de la page
        response = requests.get(url)

        if response.status_code == 200:
            # Analyser le contenu HTML avec BeautifulSoup
            soup = BeautifulSoup(response.content, 'html.parser')

            # Trouver la div avec la classe "infobox" et "title"
            infobox_div = soup.find('div', class_='infobox')
            if infobox_div:
                title_div = infobox_div.find('div', class_='title')
                if title_div:
                    # Trouver le premier h2 après title_div
                    first_h2_after_title = None
                    for sibling in title_div.next_siblings:
                        if sibling.name == 'h2':
                            first_h2_after_title = sibling
                            break

                    # Extraire les paragraphes entre title_div et le premier h2
                    paragraphs = []
                    current_element = title_div.find_next('p')
                    while current_element and current_element != first_h2_after_title:
                        paragraphs.append(current_element.get_text(strip=True))
                        current_element = current_element.find_next('p')

                    # Joindre les paragraphes en une seule chaîne de texte
                    text_content = '\n'.join(paragraphs)

                    # Afficher le contenu extrait selon l'argument print_content
                    if print_content:
                        if isinstance(print_content, bool):
                            print(f"Contenu extrait pour {keyword} :")
                            print(text_content)
                        elif isinstance(print_content, int):
                            print(f"Contenu extrait pour {keyword} (premiers {print_content} caractères) :")
                            print(text_content[:print_content])

                    # Nom du fichier d'exportation avec le chemin du dossier
                    export_filename = os.path.join(export_dir, f'Wiki_{keyword}_Description.txt')

                    # Écrire le contenu texte dans un fichier
                    with open(export_filename, 'w', encoding='utf-8') as file:
                        file.write(text_content)

                    print(f"\n Enregistrement dans {export_filename}")
                else:
                    print(f"Aucune div avec la classe 'title' trouvée dans 'infobox' pour {keyword}.")
            else:
                print(f"Aucune div avec la classe 'infobox' trouvée pour {keyword}.")
        else:
            print(f"La requête a échoué avec le code d'état : {response.status_code} pour {keyword}")

# Liste des mots à utiliser pour construire les URLs
keywords = ['Damage', 'Mods', 'Warframes'] # Ajoutez d'autres mots selon vos besoins


# Appeler la fonction avec la liste des mots et l'option d'impression

# extract_and_save_content(keywords, print_content=True)  # Imprimer tout le contenu
extract_and_save_content(keywords, print_content=100) # Imprimer les 100 premiers caractères
# extract_and_save_content(keywords, print_content=False) # Ne pas imprimer le contenu
```
